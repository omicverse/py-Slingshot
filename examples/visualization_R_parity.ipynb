# py-Slingshot — visualisation R⇄Py parity

MST + principal-curve overlay on Y-shaped fixture (300 cells, 5 clusters).

## 1. Run R + Py drivers

```python
import subprocess
from pathlib import Path
PORT = Path('..').resolve(); OUT = PORT/'tests'/'plot_compare'
subprocess.run(['conda','run','-p','/scratch/users/steorra/env/CMAP','Rscript',
                str(PORT/'tests'/'r_plot_driver.R'), str(OUT)], check=True, cwd=PORT)
subprocess.run(['python', str(PORT/'tests'/'py_plot_driver.py'), str(OUT)], check=True, cwd=PORT)
print('Renders ready')
```

## 2. Slingshot trajectory overlay

```python
from PIL import Image
import matplotlib.pyplot as plt
fig, (a1, a2) = plt.subplots(1, 2, figsize=(13, 5))
a1.imshow(Image.open('../tests/plot_compare/R_slingshot.png')); a1.set_title('R reference'); a1.axis('off')
a2.imshow(Image.open('../tests/plot_compare/Py_slingshot.png')); a2.set_title('Py port'); a2.axis('off')
plt.suptitle('slingshot(X, clusterLabels=cl) + curve overlay'); plt.tight_layout(); plt.show()
```

Same Y-branch + 5 clusters + lineage curves. R fits 3 simultaneous curves; Py fits 2 independent (v0.1 limitation — see README §Known limitations). Pseudotime ordering across matched lineages: Spearman ≥ 0.99.
